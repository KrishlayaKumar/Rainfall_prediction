```python
from sklearn import preprocessing
r_scaler = preprocessing.MinMaxScaler()
r_scaler.fit(df)
modified_data = pd.DataFrame(r_scaler.transform(df), index=df.index, columns=df.columns)
modified_data.head()
modified_data
```

```python
# Importing required libraries for feature selection
from sklearn.feature_selection import SelectKBest, chi2

# Separating features (X) and target variable (y)
# 'X' contains all columns except 'RainTomorrow' (independent variables)
X = modified_data.loc[:, modified_data.columns != 'RainTomorrow']

# 'y' contains only 'RainTomorrow' (dependent variable)
y = modified_data[['RainTomorrow']]

# Creating a feature selector that uses the Chi-Square (χ²) statistical test
# SelectKBest selects the top 'k' best features based on the highest Chi-Square scores
selector = SelectKBest(chi2, k=7)  # Selects top 7 features

# Fitting the selector to the dataset (calculating Chi-Square scores for all features)
selector.fit(X, y)

# Transforming 'X' to keep only the top 7 selected features
X_new = selector.transform(X)

# Printing the names of the selected top 7 features
print(X.columns[selector.get_support(indices=True)])
```

```python
import matplotlib.pyplot as plt

# Get Chi-Square scores for all features
chi2_scores = selector.scores_

# Create a DataFrame with feature names and their scores
feature_scores = pd.DataFrame({'Feature': X.columns, 'Chi2 Score': chi2_scores})

# Sort features by Chi2 Score (descending)
feature_scores = feature_scores.sort_values(by='Chi2 Score', ascending=False)

# Plot the top features
plt.figure(figsize=(12, 6))
plt.barh(feature_scores['Feature'], feature_scores['Chi2 Score'], color='skyblue')
plt.xlabel('Chi-Square Score')
plt.ylabel('Feature')
plt.title('Feature Importance Based on Chi-Square Test')
plt.gca().invert_yaxis()  # Invert y-axis to show top features on top
plt.show()
```
